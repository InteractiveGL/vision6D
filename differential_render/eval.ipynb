```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def create_box_plot(data1, data2, data3, data4, output_path):
    # Assuming each data variable contains four sets of data points
    # Example of data1: [ [set1], [set2], [set3], [set4] ]
    # The same applies to data2, data3, and data4

    plt.figure(figsize=(12, 6))

    gs = gridspec.GridSpec(1, 4, width_ratios=[1, 1, 1, 1])
    axs = [plt.subplot(gs[i]) for i in range(4)]

    colors = ['pink', 'lightblue', 'lightgreen', 'lightyellow']
    titles = ['Rotation error', 'Translate x error', 'Translate y error', 'Translate z error']
    data_point_labels = ['u+c', 'u+b', 'g+b', 'g+c']
    datasets = [data1, data2, data3, data4]


    for i, ax in enumerate(axs):
        # Creating boxplot for each dataset
        bplot = ax.boxplot(datasets[i], vert=True, patch_artist=True)
        ax.set_title(titles[i])

        # Set custom x-tick labels
        ax.set_xticks(np.arange(1, 5))
        ax.set_xticklabels(data_point_labels)

        # Coloring each box
        for patch in bplot['boxes']:
            patch.set_facecolor(colors[i])

        # for k in range(4):
        #     ax.text(k + 1, 1, f'{1:.2f}', 
        #             verticalalignment='top', horizontalalignment='center', 
        #             color='black', fontsize=8)

        for j, data in enumerate(datasets[i].T):
            median = np.median(data)
            ax.text(j + 1, median, f'{median:.2f}', 
                    verticalalignment='center', horizontalalignment='center', 
                    color='black', fontsize=8)

        ax.yaxis.grid(True)
        ax.set_ylabel('Observed values')

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
```

```python
create_box_plot(predict_errors.T, predict_tx_errors.T, predict_ty_errors.T, predict_tz_errors.T, data_folder / "output" / "eval" / "box_plot.png" )
```
