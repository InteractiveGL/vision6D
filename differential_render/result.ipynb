```python
import os
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
from pathlib import Path
data_path = Path(os.path.abspath("")).parent.parent / "data"
output_path = data_path / "output" / "differential_render" / "108_192"
```

```python
folderid = "2023-11-15-17-49-32" # batch size equals to 256
fold_num = 0
train_folder = f"{folderid}_train" 
valid_folder = f"{folderid}_valid_sample_best256"
```

```python
# 486_right
ossicles_name = "6602"
ossicles_side = "right"
id ="0_1"

ground_truth_image = PIL.Image.open(output_path / valid_folder / f"fold_{fold_num}" / "mapping" / "target" / f"{ossicles_name}_{ossicles_side}_ossicles_processed_{id}.png")
predict_image = PIL.Image.open(output_path / valid_folder / f"fold_{fold_num}" / "mapping" / "predict" / f"{ossicles_name}_{ossicles_side}_ossicles_processed_{id}.png")

diff = np.abs(np.array(ground_truth_image) - np.array(predict_image))

plt.imshow(diff, cmap="gray")

diff_x = PIL.Image.fromarray(diff[..., 0])
diff_x.save("diff_x.png")
```
